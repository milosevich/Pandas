```python
%matplotlib inline
```

```python
df = pd.read_csv("all_alpha_19.csv", header='infer')
df.head(1000)
```

```python
q1 = df.query("Stnd=='T3B125'")
q2 = q1.query("Fuel=='Gasoline' or Fuel=='Diesel'")
q2.head(1000)
```

```python
q3 = q2[['Model', 'Displ', 'Fuel', 'City MPG', 'Hwy MPG', 'Cmb MPG', 'Greenhouse Gas Score']].reset_index(drop=True)
q3.head()
```

```python
print(q3.dtypes)
print("=" * 50)
```

```python
q4 = q3.astype({'City MPG': float, 'Hwy MPG': float, 'Cmb MPG': float})
print(q4.dtypes)
```

```python
def mpg_to_kml(mpg):
    kml = mpg * 0.4251
    return kml
```

```python
mpg1 = q4.assign(City_kml = mpg_to_kml(q4['City MPG']))
mpg1.head()
```

```python
mpg2 = mpg1.assign(Hwy_kml = mpg_to_kml(q4['Hwy MPG']))
mpg2.head()
```

```python
mpg3 = mpg2.assign(Cmb_kml = mpg_to_kml(q4['Cmb MPG']))
mpg3.head()
```

```python
mpg3.to_csv("car_data.csv", header=True, index=True)
```
