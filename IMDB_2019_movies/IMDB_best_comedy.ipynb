```python
filt = df['Genres'].str.contains('Comedy') # grab genres
```

```python
comedy = df.loc[filt] # create mask for data, show us only cells with Comedy word
```

```python
df['IMDb Rating'].value_counts()
```

```python
# Write it to a csv, just comedy movies
comedy.to_csv('data/2019_movie_rankings_comedy.csv', header=True, index=True)
```

```python
high_rating = (comedy['IMDb Rating'] > 8.0) # filter comedy movies with rating higher than 8.0
```

```python
high_rating = comedy.loc[high_rating]
```

```python
high_rating.sort_values(by=['IMDb Rating'], ascending=False) # sort by IMDb rating
```

```python
high_rating.shape # only comedy movie with high rating
```

```python
# Write it to a csv, just comedy movies with rating higher than 8.0
high_rating.to_csv('data/2019_movie_rankings_comedy_best.csv', header=True, index=True)
```
